```python
data_list = []
for d in data:
    p = int(d.split(' ')[0])
    l = int(d.split(' points')[0].split(' ')[-1])
    if "high score" in d:
        s = int(d.split(' ')[-1])
    else:
        s = 0
    data_list.append([p,l,s])
```

```python
# Let's brute for this with some for looopsss
import numpy as np
num = 0
PLAYERS = data_list[num][0]
LAST_MARBLE = data_list[num][1]

board = np.zeros((LAST_MARBLE))
prev_board = np.zeros((2))

player_scores = np.zeros((PLAYERS))

current_position = 0
num_marbles = 1

def update_position(prev_position, num_marbles):
    curr_position = prev_position + 2
    
    if curr_position >= num_marbles:
        curr_position = 1
    return curr_position

def remove_position(position, num_marbles):
    remove_position = position - 7
    if remove_position < 0:
        remove_position += num_marbles 
    return remove_position    
    
for marble in range(LAST_MARBLE):
    if marble == 0.0:
        num_marbles = 1
    elif marble == 1.0:
        num_marbles = 2
        current_position = 1%num_marbles
        player = (marble-1)%PLAYERS
        
        prev_board[current_position] = marble
    
    else:
        curr_board = prev_board[:num_marbles+1]
        curr_board = np.append(curr_board, 0)
        points = 0
        player = (marble-1)%PLAYERS
        
        if marble%20000==0:
            print (marble)
        
        if (marble%23) == 0:
            points = marble
            
            current_position = remove_position(current_position, num_marbles)
            
            points += curr_board[current_position]
            
            curr_board[current_position] = 0
            curr_board[current_position:] = np.roll(curr_board[current_position:], -1)
            
            player_scores[player] += points
            num_marbles -= 1
        
        else:
            num_marbles += 1
            current_position = update_position(current_position, num_marbles)

            curr_board[current_position:] = np.roll(curr_board[current_position:], 1)
            curr_board[current_position] = marble
        prev_board = curr_board
        
    #print prev_board
        
    #print (player, board, current_position, num_marbles)
print ("Done")
```
